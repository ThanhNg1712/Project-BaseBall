```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
games = pd.read_csv('/Users/thanhnguyen/Desktop/Data_analysis/DE_k2/project1_baseball/games/games.csv')
games
```

```python
#task1
plays = games[games['type'] == 'play']
plays.head()
```

```python
#task2
strike_outs = plays[plays['event'].str.contains('K')]
strike_outs.head()
```

```python
#task3

strike_outs = strike_outs.groupby(['year', 'game_id']).size()
#task4
strike_outs = strike_outs.reset_index(name='strike_outs')
strike_outs.head()
```

```python
#task5
strike_outs = strike_outs.loc[:,['year','strike_outs']].apply(pd.to_numeric)
```

```python
#task6
plt.figure(figsize=(15,8))
plt.scatter(strike_outs['year'],strike_outs['strike_outs'])
plt.legend(['Strike Outs'])
plt.show()
```
