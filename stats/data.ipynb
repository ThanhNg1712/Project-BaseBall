```python

games_dir = os.path.join(os.getcwd(), 'games')
game_files = glob.glob(os.path.join(games_dir, '*.EVE'))
```

```python
game_files.sort()
#game_files
```

```python
game_frames =[]


for game_file in game_files:
    game_frame = pd.read_csv(game_file, names = ['type','multi2', 'multi3', 'multi4', 'multi5', 'multi6', 'event'])
    
    
    #add to the game_frames list
    
    game_frames.append(game_frame)
    
type(game_frames)
game_frames
```

```python
#task 7
games = pd.concat(game_frames)
games.head()
```

```python
#task 8
#clean ?? value in multi5

games.loc[games.multi5 == '??', 'multi5'] = ''
```

```python
#task 9
#extract identifiers
identifiers = games['multi2'].str.extract(r'(.LS(\d{4})\d{5})')
identifiers.head()
# null_count = identifiers[1].isnull().sum()
# print(null_count)
```

```python
#task10
identifiers = identifiers.fillna(method='ffill')
identifiers.head()
```

```python
#task 11
identifiers.columns=(['game_id','year'])
#task 12
games = pd.concat([games, identifiers], axis=1, sort=False)
games
```

```python
#task13
games = games.fillna('')
games.head()
```

```python
#task14
games['type'] = pd.Categorical(games.loc[:, 'type'])
games.type.dtypes
```

```python
#task15
games.head()
```
